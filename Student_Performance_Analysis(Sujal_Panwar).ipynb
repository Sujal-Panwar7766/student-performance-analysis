## Data Loading
The dataset was loaded using pandas. It contains student information like grades, study time, and gender.

## Analysis
- Calculated average final grade (G3)

## Visualization
- Histogram shows distribution of grades
- Scatter plot shows relation between study time and performance
- Bar chart compares male vs female scores

1.  Histogram shows distribution of grades

```python
plt.hist(df['G3'], bins=10)
plt.title("Distribution of Final Grades")
plt.xlabel("Grades")
plt.ylabel("Frequency")
plt.show()
```

2. Scatterplot: study time vs grades.

```python
plt.scatter(df['studytime'], df['G3'])
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.title("Study Time vs Grade")
plt.show()
```

3. Bar chart: male vs female average score.

```python
df.groupby('sex')['G3'].mean().plot(kind='bar')
plt.title("Average Score by Gender")
plt.show()
```

## Advanced Visualizations
These visualizations provide deeper insights into student performance using advanced plotting techniques.

1. Correlation Heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

2. Box Plot (Detect Outliers)

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['G3'])
plt.title("Box Plot of Final Grades")
plt.show()
```

## Conclusion
- Students with higher study time tend to perform better
